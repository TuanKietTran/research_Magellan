````python
DBLP2 = em.read_csv_metadata('DBLP2.csv', key='id')
acm, dblp2 = em.down_sample(ACM, DBLP2, size=1000, y_param=1, show_progress=False)
acm = acm.reset_index().drop(['index'], axis=1)
dblp2 = dblp2.reset_index().drop(['index'], axis=1)
em.set_property(acm, 'key', 'id')
em.get_key(acm)
em.set_property(dblp2, 'key', 'id')
em.get_key(dblp2)

````

Phần blocker chính (block year, title)

````python
ob = em.OverlapBlocker()
block_set1 = ob.block_tables(acm, dblp2, 'title', 'title', overlap_size=3, l_output_attrs=['title', 'authors', 'venue', 'year'], r_output_attrs=['title', 'authors', 'venue', 'year'], allow_missing=True )
eb = em.AttrEquivalenceBlocker()
block_set1 = eb.block_candset(block_set1, 'year', 'year')
def match_blocker1(ltuple, rtuple):
    l_ = ltuple['venue'].lower().strip()
    r_ = rtuple['venue'].lower().strip()

    l_venue = ['ACM Transactions on Database Systems (TODS)', 'ACM SIGMOD Record', 'International Conference on Management of Data', 'Very Large Data Bases', 'The VLDB Journal &mdash; The International Journal on Very Large Data Bases']
    r_venue = ['ACM Trans. Database Syst.', 'SIGMOD Record', 'SIGMOD Conference', 'VLDB', 'VLDB J.']

    for i in range(5):
    
      if l_ == l_venue[i].lower().strip() and r_ == r_venue[i].lower().strip():
        return False
    
    return True

bb1 = em.BlackBoxBlocker()
bb1.set_black_box_function(match_blocker1)
block_set1 = bb1.block_candset(block_set1)

def match_blocker2(ltuple, rtuple):
    l_ = ltuple['title']
    r_ = rtuple['title']
    lengl = len(l_.split())
    lengr = len(r_.split())

    if (lengl in [1,2] or lengr in [1,2]):
      return False
    else:
      return True

block_set2 = ob.block_tables(acm, dblp2, 'title', 'title', l_output_attrs=['title', 'authors', 'venue', 'year'], r_output_attrs=['title', 'authors', 'venue', 'year'], overlap_size= 1)
block_set2 = eb.block_candset(block_set2, 'year', 'year')
bb = em.BlackBoxBlocker()
bb.set_black_box_function(match_blocker2)
block_set2 = bb.block_candset(block_set2)
bb.set_black_box_function(match_blocker1)
block_set2 = bb.block_candset(block_set2)


````

````python
combine_blocked_set = em.combine_blocker_outputs_via_union([block_set1, block_set2])
len(combine_blocked_set)
````

````python
corres = [('title', 'title'), ('authors', 'authors'), ('year', 'year'), ('venue', 'venue')]
debug_set = em.debug_blocker(combine_blocked_set, acm, dblp2, attr_corres=corres, output_size=200)
debug_set.head()
````

````python
S = em.sample_table(combine_blocked_set, 600)
````

Gắn nhãn từ file gt

````python
G = S
lst = []
for row in S.itertuples():
  tab = gt[gt['idACM'] == row[2]]
  flag = False
  for r in tab.itertuples():
    if row[3] == r[1]:
      lst.append(1) 
      flag = True
      break
  if not flag:
    lst.append(0)

G['gold'] = lst
G.head(20)
````

````python
train_test = em.split_train_test(G, train_proportion=0.7)
train_set = train_test['train']
test_set = train_test['test']
````

Thực hiện chọn các feature dùng cho 2 tập

````python
match_t = em.get_tokenizers_for_matching()
match_s = em.get_sim_funs_for_matching()
atypes1 = em.get_attr_types(acm) # don't need, if atypes1 exists from blocking step
atypes2 = em.get_attr_types(dblp2) # don't need, if atypes2 exists from blocking step
match_c = em.get_attr_corres(acm, dblp2)
match_f = em.get_features(acm, dblp2, atypes1, atypes2, match_c, match_t, match_s)
````

````python
train_set_vecs = em.extract_feature_vecs(train_set, 
                            feature_table=match_f, 
                            attrs_after='gold',
                            show_progress=False)
````

Thay NaN bằng mean (nếu có)

````python
print(train_set_vecs['authors_authors_jac_qgm_3_qgm_3'].mean())
print(train_set_vecs['authors_authors_cos_dlm_dc0_dlm_dc0'].mean())
print(train_set_vecs['authors_authors_mel'].mean())
print(train_set_vecs['authors_authors_lev_dist'].mean())
print(train_set_vecs['authors_authors_lev_sim'].mean())
````

````python
train_set_vecs['authors_authors_jac_qgm_3_qgm_3'] = train_set_vecs['authors_authors_jac_qgm_3_qgm_3'].fillna(value=train_set_vecs['authors_authors_jac_qgm_3_qgm_3'].mean(), inplace=False)
train_set_vecs['authors_authors_cos_dlm_dc0_dlm_dc0'] = train_set_vecs['authors_authors_cos_dlm_dc0_dlm_dc0'].fillna(value=train_set_vecs['authors_authors_cos_dlm_dc0_dlm_dc0'].mean(), inplace=False)
train_set_vecs['authors_authors_mel'] = train_set_vecs['authors_authors_mel'].fillna(value=train_set_vecs['authors_authors_mel'].mean(), inplace=False)
train_set_vecs['authors_authors_lev_dist'] = train_set_vecs['authors_authors_lev_dist'].fillna(value=train_set_vecs['authors_authors_lev_dist'].mean(), inplace=False)
train_set_vecs['authors_authors_lev_sim'] = train_set_vecs['authors_authors_lev_sim'].fillna(value=train_set_vecs['authors_authors_lev_sim'].mean(), inplace=False)
````

````python
dt = em.DTMatcher(name='DecisionTree', random_state=0)
svm = em.SVMMatcher(name='SVM', random_state=0)
rf = em.RFMatcher(name='RF', random_state=0)
lg = em.LogRegMatcher(name='LogReg', random_state=0)
ln = em.LinRegMatcher(name='LinReg')
nb = em.NBMatcher(name='NaiveBayes')
result = em.select_matcher([dt, rf, svm, ln, lg, nb], table=train_set_vecs, 
        exclude_attrs=['_id', 'ltable_id', 'rtable_id', 'gold'],
        k=5,
        target_attr='gold', metric_to_select_matcher='precision', random_state=0)
result['cv_stats']
````
